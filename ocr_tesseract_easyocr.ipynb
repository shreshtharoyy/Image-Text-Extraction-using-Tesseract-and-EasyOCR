PyTesseract

```python
import pytesseract
from PIL import Image

img_path = '/content/ocr-receipts-text-detection/images/0.jpg'

text = pytesseract.image_to_string(Image.open(img_path), lang='eng')
print(text)
```

EasyOCR

```python
from easyocr import Reader

reader = Reader(['en'])
easytext = ''

result = reader.readtext(img_path)

for res in result:
  easytext += res[1]
  easytext += ' '

easytext = easytext[:-1]
print(easytext)
```
